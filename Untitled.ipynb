```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot bar graph for specified rows
def plot_bar_for_rows(df, start, end, title):
    selected_rows = df.iloc[start:end]
    plt.figure(figsize=(10, 6))
    plt.bar(selected_rows.index, selected_rows['Percent'], color='skyblue')
    plt.title(title)
    plt.xlabel('Index')
    plt.ylabel('Percent')
    plt.grid(axis='y')
    plt.show()

# Plot bar graphs for the specified rows
plot_bar_for_rows(df, 171, 191, 'Percentage Response from Rows 172 to 191')
plot_bar_for_rows(df, 266, 286, 'Percentage Response from Rows 267 to 286')
plot_bar_for_rows(df, 346, 362, 'Percentage Response from Rows 347 to 362')
plot_bar_for_rows(df, 422, 438, 'Percentage Response from Rows 423 to 438')
plot_bar_for_rows(df, 498, 514, 'Percentage Response from Rows 499 to 514')
```

```python

import seaborn as sns


# Extracting the specified rows to create a single dataframe for correlation analysis
rows_for_correlation = pd.concat([
    df.iloc[76:96],
    df.iloc[171:191],
    df.iloc[266:286],
    df.iloc[346:362],
    df.iloc[422:438],
    df.iloc[498:514]
])

# Selecting only numeric columns
numeric_columns = rows_for_correlation.select_dtypes(include='number')

# Creating the correlation matrix
correlation_matrix = numeric_columns.corr()

# Visualizing the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
